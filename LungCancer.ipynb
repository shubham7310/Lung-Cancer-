```python
# Import necessary libraries
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from imblearn.over_sampling import SMOTE

# Step 1: Feature Scaling using MinMaxScaler to avoid negative values
scaler = MinMaxScaler()
X = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']
X_scaled = scaler.fit_transform(X)

# Step 2: Balancing the Classes using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Step 3: Splitting Data into Train and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.25, random_state=42)

# Step 4: Hyperparameter Tuning Functions
def tune_logistic_regression(X_train, y_train):
    params = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l2']}
    grid = GridSearchCV(LogisticRegression(random_state=42), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_decision_tree(X_train, y_train):
    params = {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 10, 20],
        'criterion': ['gini', 'entropy']
    }
    grid = GridSearchCV(DecisionTreeClassifier(random_state=42), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_knn(X_train, y_train):
    params = {'n_neighbors': [3, 5, 7, 9], 'metric': ['minkowski', 'euclidean']}
    grid = GridSearchCV(KNeighborsClassifier(), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_svc(X_train, y_train):
    params = {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': ['scale', 'auto']
    }
    grid = GridSearchCV(SVC(random_state=42), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_random_forest(X_train, y_train):
    params = {
        'n_estimators': [100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 10],
        'criterion': ['gini', 'entropy']
    }
    grid = GridSearchCV(RandomForestClassifier(random_state=42), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_gradient_boost(X_train, y_train):
    params = {
        'learning_rate': [0.01, 0.1, 0.2],
        'n_estimators': [100, 200],
        'max_depth': [3, 5, 7]
    }
    grid = GridSearchCV(GradientBoostingClassifier(random_state=42), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_xgboost(X_train, y_train):
    params = {
        'learning_rate': [0.01, 0.1, 0.2],
        'n_estimators': [100, 200],
        'max_depth': [3, 5, 7]
    }
    grid = GridSearchCV(XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_mlp(X_train, y_train):
    params = {
        'hidden_layer_sizes': [(50,), (100,), (50, 50)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'lbfgs']
    }
    grid = GridSearchCV(MLPClassifier(random_state=42), params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

# Step 5: Training and Evaluating Models
models = {
    'Logistic Regression': tune_logistic_regression(X_train, y_train),
    'Decision Tree': tune_decision_tree(X_train, y_train),
    'KNN': tune_knn(X_train, y_train),
    'Gaussian Naive Bayes': GaussianNB(),
    'Multinomial Naive Bayes': MultinomialNB(),  # Should work now with MinMaxScaler
    'SVC': tune_svc(X_train, y_train),
    'Random Forest': tune_random_forest(X_train, y_train),
    'Gradient Boosting': tune_gradient_boost(X_train, y_train),
    'XGBoost': tune_xgboost(X_train, y_train),
    'MLP Classifier': tune_mlp(X_train, y_train)
}

# Evaluate all models and print classification report
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))
```

```python
# Import necessary libraries
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from imblearn.over_sampling import SMOTE
import numpy as np

# Step 1: Feature Scaling using MinMaxScaler to avoid negative values
scaler = MinMaxScaler()
X = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']
X_scaled = scaler.fit_transform(X)

# Step 2: Feature Selection
selector = SelectKBest(f_classif, k=10)
X_selected = selector.fit_transform(X_scaled, y)

# Step 3: Balancing the Classes using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_selected, y)

# Step 4: Splitting Data into Train and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.25, random_state=42)

# Step 5: Hyperparameter Tuning Functions with expanded search space
def tune_logistic_regression(X_train, y_train):
    params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], 'penalty': ['l1', 'l2'], 'solver': ['liblinear', 'saga']}
    grid = RandomizedSearchCV(LogisticRegression(random_state=42, max_iter=10000), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_decision_tree(X_train, y_train):
    params = {
        'max_depth': [None, 10, 20, 30, 40, 50, 70, 100],
        'min_samples_split': [2, 5, 10, 20, 30],
        'criterion': ['gini', 'entropy'],
        'min_samples_leaf': [1, 2, 5]
    }
    grid = RandomizedSearchCV(DecisionTreeClassifier(random_state=42), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_knn(X_train, y_train):
    params = {'n_neighbors': range(3, 15, 2), 'metric': ['minkowski', 'euclidean', 'manhattan']}
    grid = RandomizedSearchCV(KNeighborsClassifier(), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_svc(X_train, y_train):
    params = {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
        'gamma': ['scale', 'auto']
    }
    grid = RandomizedSearchCV(SVC(random_state=42, probability=True), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_random_forest(X_train, y_train):
    params = {
        'n_estimators': [100, 200, 300, 500],
        'max_depth': [None, 10, 20, 30, 50, 70],
        'min_samples_split': [2, 5, 10, 20],
        'min_samples_leaf': [1, 2, 5],
        'criterion': ['gini', 'entropy']
    }
    grid = RandomizedSearchCV(RandomForestClassifier(random_state=42), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_gradient_boost(X_train, y_train):
    params = {
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'n_estimators': [100, 200, 300, 500],
        'max_depth': [3, 5, 7, 9]
    }
    grid = RandomizedSearchCV(GradientBoostingClassifier(random_state=42), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_xgboost(X_train, y_train):
    params = {
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'n_estimators': [100, 200, 300, 500],
        'max_depth': [3, 5, 7, 9],
        'subsample': [0.7, 0.8, 1.0]
    }
    grid = RandomizedSearchCV(XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

def tune_mlp(X_train, y_train):
    params = {
        'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'lbfgs'],
        'alpha': [0.0001, 0.001, 0.01]
    }
    grid = RandomizedSearchCV(MLPClassifier(random_state=42, max_iter=10000), params, cv=5, n_jobs=-1, n_iter=50)
    grid.fit(X_train, y_train)
    return grid.best_estimator_

# Step 6: Training and Evaluating Models
models = {
    'Logistic Regression': tune_logistic_regression(X_train, y_train),
    'Decision Tree': tune_decision_tree(X_train, y_train),
    'KNN': tune_knn(X_train, y_train),
    'Gaussian Naive Bayes': GaussianNB(),
    'Multinomial Naive Bayes': MultinomialNB(),  # Should work now with MinMaxScaler
    'SVC': tune_svc(X_train, y_train),
    'Random Forest': tune_random_forest(X_train, y_train),
    'Gradient Boosting': tune_gradient_boost(X_train, y_train),
    'XGBoost': tune_xgboost(X_train, y_train),
    'MLP Classifier': tune_mlp(X_train, y_train)
}

# Step 7: Initialize lists to store results
accuracy_scores = {}
cross_val_scores = {}

# Evaluate all models, print classification report, and store accuracy and cross-validation results
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Calculate accuracy and cross-validation
    accuracy = accuracy_score(y_test, y_pred)
    cross_val = cross_val_score(model, X_resampled, y_resampled, cv=5).mean()

    # Store results
    accuracy_scores[model_name] = accuracy
    cross_val_scores[model_name] = cross_val

    # Print classification report
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))

# Step 8: Print accuracy and cross-validation scores summary
print("\nModel Performance Summary:")
print("Model\t\t\tAccuracy\tCross-Validation")
for model_name in models.keys():
    print(f"{model_name}\t\t{accuracy_scores[model_name]:.4f}\t\t{cross_val_scores[model_name]:.4f}")

# Step 9: Ensemble Stacking/Voting
voting_clf = VotingClassifier(
    estimators=[
                ('random_forest', models['Random Forest']),
        ('gradient_boost', models['Gradient Boosting']),
        ('svc', models['SVC']),
        ('xgboost', models['XGBoost']),
    ],
    voting='soft',  # Use soft voting for probabilities
    weights=[2, 2, 1, 2]  # Adjust weights based on individual model performance
)

# Fit and evaluate the ensemble model
voting_clf.fit(X_train, y_train)
y_voting_pred = voting_clf.predict(X_test)

# Calculate accuracy and cross-validation for the ensemble model
voting_accuracy = accuracy_score(y_test, y_voting_pred)
voting_cross_val = cross_val_score(voting_clf, X_resampled, y_resampled, cv=5).mean()

# Print ensemble model performance
print("\nEnsemble Voting Classifier Performance:")
print(f"Accuracy: {voting_accuracy:.4f}")
print(f"Cross-Validation Score: {voting_cross_val:.4f}")

# Summary of Ensemble and Other Models
print("\nOverall Performance Summary:")
print(f"Voting Classifier Accuracy: {voting_accuracy:.4f}")
print(f"Voting Classifier Cross-Validation Score: {voting_cross_val:.4f}")
print("\nIndividual Model Performance Summary:")
print("Model\t\t\tAccuracy\tCross-Validation")
for model_name in models.keys():
    print(f"{model_name}\t\t{accuracy_scores[model_name]:.4f}\t\t{cross_val_scores[model_name]:.4f}")

```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define data for model performance metrics
data = {
    "Model": [
        "KNN", "Gaussian Naive Bayes", "Multinomial Naive Bayes",
        "SVC", "Random Forest", "Gradient Boosting",
        "XGBoost", "MLP Classifier"
    ],
    "Precision_0": [0.95, 0.92, 0.75, 0.95, 0.95, 0.95, 0.95, 0.94],
    "Recall_0": [0.97, 0.90, 0.77, 0.97, 0.97, 0.95, 0.98, 0.98],
    "F1_Score_0": [0.96, 0.91, 0.76, 0.96, 0.96, 0.95, 0.97, 0.96],
    "Precision_1": [0.97, 0.90, 0.76, 0.97, 0.97, 0.95, 0.98, 0.98],
    "Recall_1": [0.95, 0.92, 0.75, 0.95, 0.95, 0.95, 0.95, 0.93],
    "F1_Score_1": [0.96, 0.91, 0.75, 0.96, 0.96, 0.95, 0.97, 0.96]
}

# Create DataFrame
df_metrics = pd.DataFrame(data)

# Define the number of metrics and the width of each bar
metrics = ["Precision", "Recall", "F1_Score"]
width = 0.2  # Width of each bar

# Create grouped bar chart
fig, ax = plt.subplots(figsize=(14, 8))

# Set positions for each group and each bar
x = np.arange(len(df_metrics["Model"]))
for i, metric in enumerate(metrics):
    ax.bar(x - width + i*width, df_metrics[f"{metric}_0"], width, label=f"{metric} (Class 0)")
    ax.bar(x + i*width, df_metrics[f"{metric}_1"], width, label=f"{metric} (Class 1)", alpha=0.7)

# Labeling the chart
ax.set_xlabel("Model")
ax.set_ylabel("Score")
ax.set_title("Precision, Recall, and F1 Score for Each Model (Class 0 and Class 1)")
ax.set_xticks(x)
ax.set_xticklabels(df_metrics["Model"], rotation=45, ha="right")
ax.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
